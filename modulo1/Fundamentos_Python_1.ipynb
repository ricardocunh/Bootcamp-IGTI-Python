# **Python "Stack"(Pilha)**

* Permite adicionar e remover elementos
* Elementos são adicionados e retirados do top da pilha

```
# Função que inverte o valor da String passada
def inverte_string(valor_string):
  resultado = " "
  for caracter in valor_string:
    resultado = caracter + resultado

  return resultado
```

```
# Utilizando a função definida

string = input("Entre com alguma string: ")
while string.strip()!="":
  print("O inverso de ",string, "é ", inverte_string(string))
  string=input("Entre com outra string ou pressione ENTER para sair: ")
```

# **Predicate Functions**

```
# função que retorna verdadeiro ou falso - verifica se um número x é divisível pelo número de y

def verifica_divisibilidade(x,y):
  x_divisivel_y = False

  if x % y ==0:
    x_divisivel_y = True

  return x_divisivel_y
```

```
x = int(input("Entre com um valor inteiro: "))
y = int(input("Entre com outro valor inteiro: "))

if verifica_divisibilidade(x,y):
  print(x, " é divisível por ", y)
else:
  print(x, " não é divisível por ", y)
```

# **Desenvolvimento TOP-DOWN x BOTTOM-UP**

**TOP-DOWN**

* O Programa completo é dividido em partes menores;
* Normalmente, utilizado em programação estruturada;
* Cada parte é programada separadamente o que gera redundância;
* Baseado em "Decomposição;


**BOTTOM-UP**
* Partes menores são resolvidas e integradas para o programa maior;
* Normalmente, utilizada em programação orientada a objetos;
* Redundância é minimizada devido a técnica como o encapsulamento;
* Baseado em "Composição"

```
# construindo o programa via bottom_up
def verifica_divisibilidade(x,y):
  return int(y) % int(x)==0
```

```
# em um modelo de bottom-up
#definindo uma função que retorna a lista com os valores da lista que nsão divisores de x
def elementos_divisiveis(x,lista):
  resultado=[]

  for elemento in lista:
    if verifica_divisibilidade(elemento,x):
      resultado.append(elemento)
  return resultado
```
