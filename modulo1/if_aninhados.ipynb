**If com condições mais elaboradas**

```
# if com condicional or

altura = int(input("Entre com sua altura (em centimentros): "))
if(altura<150) or (altura>180):
  print("Você não pode brincar")
else:
  print("Você pode brincar")
```

```
# Adivinhando resultados da divisão
numerador = float(input("Entre com o numerador: "))
denominador=float(input("Entre com o denominador: "))

chute=float(input("Entre com seu chute para o resultado: "))

resultado=numerador/denominador

maior=abs(resultado)


#verifica se o chute é até 1/10 do real
if abs((chute - resultado)/maior) <=0.1:
  print("Seu chute está correto. Parabéns!")
else:
  print("Que pena, você errou. O valor correto é {}".format(resultado))
```

```
i = 10
if (i == 10):
  #Primeiro if
  if(i < 15):
    print("i é maior que 15")
  # o if aninhado só será executado se o anterior for executado
  if (i< 12):
    print("i também é menor que 12")
  else:
    print("i é maior que 15")
  print("Continua o programa...")
```

```
x = float(input("Entre com o primeiro número: "))
y = float(input("Entre com o segundo número: "))

print('(1) - Soma os dois números')
print('(2) - Subtrai os dois números')
print('(3) - Multiplica os dois números')
print('(4) - Divide os dois números')

escolha=int(input("Escolha uma das opções: "))

print("O Resultado é: ", end="")

if escolha==1:
  print(x+y)
else:
  if (escolha==2):
    print(x-y)
  else:
    if (escolha==3):
      print(x*y)
    else:
      if (escolha==4):
        print(x/y)
      else:
        print("Opção inválida")
```

**Encontrando o maior entre 3 números**

```
x = int ( input("Entre com o primeiro número inteiro: "))
y = int ( input("Entre com o segundo número inteiro: "))
z = int ( input("Entre com o terceiro número inteiro: "))

if y > x:
  if z > y:
    print(z, " é o maior número.")
  else:
    print(y, " é o maior número.")
else:
  if z > x:
    print(z, " é o maior número.")
  else:
    print(x, " é o maior número.")
print("FIM")
```

**Forma mais inteligente**

```
x = int(input("Entre com o primeiro número inteiro: "))
y = int(input("Entre com o segundo número inteiro: "))
z = int(input("Entre com o terceiro número inteiro: "))

#Primeiro chute
maior_numero = x
if y > maior_numero:
  #ajusta o primeiro chute
  maior_numero = y
if z > maior_numero:
  #ajusta o segundo chute
  maior_numero = z
print(maior_numero, " é o maior número.")
print("FIM")
```

```
# lidando com erros - verificando se um número é divisível por outro

numerador=int(input("Entre com o numerador"))
denominador=int(input("Entre com o denominador"))

if numerador % denominador ==0:
  print("O numerador é divisível pelo denominador.")
else:
  print("A divisão não é inteira")
```

```
# Exemplo de adivinhação com exceções
try:
  numerador=int(input("Entre com o numerador"))
except ValueError: #captura apenas exceções do tipo ValueErrors
  print("Você não entrou com um número inteiro")
  exit(0)

try:
  denominador=int(input("Entre com o denominador"))
except ValueError:
  print("Você não entrou com um número interio")
  exit(0)

try:
  if numerador % denominador ==0:
    print("O numerador é divisível pelo denominador")
  else:
    print("A Divisão não é inteira")
except ZeroDivisionError: #captura divisões por zero
  print("Denominador não pode ser zero")    
```

```
idades = {'Ricardo': 35, 'Gisele': 89, 'Nicolly': 9}
pessoa = input("Quero saber a idade de: ")
idade = idades.get(pessoa)

if idade:
  print(f"{pessoa} tem {idade} anos de idade.")
else:
  print(f"{pessoa} com idade desconhecida.")
```

```
idades = {'Ricardo': 35, 'Gisele': 89, 'Nicolly': 9}
pessoa = input("Quero saber a idade de: ")
idade = idades.get(pessoa)

try:
  print(f"{pessoa} tem {idades[pessoa]} anos de idade.")
except KeyError:
  print(f"{pessoa} com idade desconhecida.")
```
