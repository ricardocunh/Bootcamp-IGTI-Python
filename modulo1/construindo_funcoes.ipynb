# **A Função MAIN**

**Python não precisa, mas é legal!**

```
# a função main básica em python

def main():
  print("Ola Mundo!")

if __name__=="__main__":
  main()
```

```
# identificando o contexto em que estamos trabalhando
print("A variável __name__ informa qual é o contexto em que estamos executando o arquivo.")
print("O valor  da variável __name__ é:", repr(__name__))
```

```
# utilizando a função main para deixar o código "mais limpo"
from time import sleep

def process_data(data):
  print("Iniciando o processo de algum dado..")
  modified_data = data + " o dado foi modificado"

  sleep(3)
  print("Processamento do dado terminado.")
  return modified_data

def read_data_from_web():
  print("Lendo dados da Web")
  data = "Dados da web lidos"
  return data

def write_data_to_database(data):
  print("Armazenando dados em um database")
  print(data)

def main():
  data = read_data_from_web()
  modified_data = process_data(data)
  write_data_to_database(modified_data)

if __name__=="__main__":
  main()
```

# **Keywords Arguments em Python**

```
# exemplo de função

def concatena_nome_mensagem(nome, mensagem): # a função possui 2 parâmetros
    print("Olá", nome + '! ' + mensagem)
  
# chamando a função corretamente
concatena_nome_mensagem("Maria", "Bom dia")
```

**Em Python existem 3 formas diferentes de definir uma quantidade variável de argumentos**

```
# exemplo de função

def concatena_nome_mensagem(nome, mensagem='Bom dia!'):
  print("Olá", nome + '! ' + mensagem)

#chamando a função com outros valores
concatena_nome_mensagem("Maria", "Boa Noite!")
```

```
#chamando a função com apenas 1 argumento

concatena_nome_mensagem('Maria')
```

**2. Via posição dos argumentos**

```
# exemplo de função
def concatena_nome_mensagem(nome, mensagem):
  print("Olá", nome + '! ' + mensagem)

concatena_nome_mensagem("Maria", "Bom dia")
```

**3. Através de argumentos arbitrários**

```
# exemplo de função (o * significa que será passado uma lista de nomes)
def print_nomes_dos_usuarios(*nomes): # a função possui N parâmetros
    
    for nome in nomes:
      print("Olá, " +nome+"! Bom dia!")
    
# Chamando a função corretamente
print_nomes_dos_usuarios("Maria", "Antônio", "José", "Ricardo")
```

```
# como o python faz isso
def minha_funcao(**kwargs):
  print(str(kwargs))

minha_funcao(a=12, b="abc")

{'a': 12, 'b': 'abc'}
```

```
# definição com a posição de apenas 2 valores
def somatorio(a, b):
  print("Somatório é: ", a+b )

somatorio(21,5)
```

```
# definicao com o *args
def somatorio(*args):
  total = 0;
  for arg in args:
    total += arg
  print("Somatorio é :",total)
```

```
somatorio(21,5,6)
```
